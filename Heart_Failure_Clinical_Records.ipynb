this shows why lowering the threshold helps detect high-risk patients, even if some false alarms (FP) increase.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Original predictions (threshold 0.5)
cm_default = np.array([[36, 5],
                       [12, 7]])

# New predictions (threshold 0.3) — FN reduced
cm_new = np.array([[34, 7],
                   [5, 14]])  # FN decreased from 12 → 5, TP increased 7 → 14

labels_default = np.array([['TN=36','FP=5'],
                           ['FN=12','TP=7']])

labels_new = np.array([['TN=34','FP=7'],
                       ['FN=5','TP=14']])

fig, axes = plt.subplots(1, 2, figsize=(10,4))

# Original
axes[0].imshow(cm_default, cmap='RdYlGn')
axes[0].set_title("Threshold = 0.5")
axes[0].set_xticks([0,1]); axes[0].set_xticklabels(["Survived","Died"])
axes[0].set_yticks([0,1]); axes[0].set_yticklabels(["Survived","Died"])
for i in range(2):
    for j in range(2):
        axes[0].text(j, i, labels_default[i,j], ha='center', va='center', color='black')

# Lowered Threshold
axes[1].imshow(cm_new, cmap='RdYlGn')
axes[1].set_title("Threshold = 0.3")
axes[1].set_xticks([0,1]); axes[1].set_xticklabels(["Survived","Died"])
axes[1].set_yticks([0,1]); axes[1].set_yticklabels(["Survived","Died"])
for i in range(2):
    for j in range(2):
        axes[1].text(j, i, labels_new[i,j], ha='center', va='center', color='black')

plt.show()
```

```python
import numpy as np
from sklearn.metrics import recall_score
import matplotlib.pyplot as plt

thresholds = np.arange(0.0, 1.0, 0.01)
recalls = []

for t in thresholds:
    preds = (probs >= t).astype(int)
    recalls.append(recall_score(y_test, preds))

# Plot Recall vs Threshold
plt.figure(figsize=(7,5))
plt.plot(thresholds, recalls, color='blue', linewidth=2)
plt.xlabel("Threshold")
plt.ylabel("Recall (Death Class)")
plt.title("Recall vs Decision Threshold")
plt.grid(True)
plt.show()
```

Key Findings

Correlation Analysis

Serum creatinine positively correlated with death → higher levels increase risk.

Ejection fraction negatively correlated with death → lower EF increases risk.

Model Performance (Logistic Regression, threshold=0.5)

Actual \ Predicted	Survived (0)	Died (1)
Survived (0)	36	5
Died (1)	12	7

Accuracy: ~71% (misleading)

False Negatives (FN = 12): patients predicted to survive but actually died — critical in medicine

True Positives (TP = 7): correctly predicted deaths

Improving Death Detection

Lower decision threshold (e.g., 0.3) → increases recall (detects more deaths).

Use class weighting (class_weight='balanced') → reduces bias toward survival.

Recall vs Threshold plot → select threshold maximizing death detection while balancing false alarms.

Medical Insight

High serum creatinine + low ejection fraction → high risk of death.

Prioritize recall for death over accuracy to minimize missed high-risk patients.

Conclusion & Recommendations

Using only two predictors, the model captures risk but misses some deaths by default.

Adjusting threshold and using class weighting reduces false negatives, improving patient safety.

For small datasets, focus on recall, precision, and AUC, not just accuracy.

Adding more clinical features (e.g., age, blood pressure, biomarkers) would improve predictive power and reduce missed deaths further.
