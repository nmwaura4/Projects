```python
# Bank ATM

def ATM_Banking():

    balance = 3000

    while True:
        print("------ ATM MENU -----")
        print("1. Withdraw Funds")
        print("2. Deposit Funds")
        print("3. Check Balance")
        print("4. Exit")

        user_choice = input("Enter choice (1-4): ")

        if user_choice == "1":
            amount = int(input("Enter amount to withdraw: "))

            if amount > balance:
                print("Insufficient balance")

            elif amount <= 0:
                print("Invalid amount")

            else:
                balance -= amount
                print("Withdrawal successful. Please collect your cash")

        elif user_choice == "2":
            amount = int(input("Enter amount you want to deposit: "))

            if amount > 0:
                balance += amount
                print("Amount deposited successfully:", amount)

            else:
                print("Invalid amount")

        elif user_choice == "3":
            print("Your balance is:", balance)

        elif user_choice == "4":
            print("Thank you for using the ATM")
            break

        else:
            print("Invalid choice! Please try again")


ATM_Banking()
```
